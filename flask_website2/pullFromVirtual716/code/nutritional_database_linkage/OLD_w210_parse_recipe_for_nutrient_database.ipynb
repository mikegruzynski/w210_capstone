```python
def preprocess_initial(ingredient):
    ingredient = ingredient.lower()
    ingredient_split = ingredient.split()

    itr = 0
    while itr < len(ingredient_split):
        if ingredient_split[itr] in food_format:                
            ingredient_split[itr] = ""
        if ingredient_split[itr] in food_size:                
            ingredient_split[itr] = ""
        itr += 1

    ingredient_split = filter(None, ingredient_split)

    ingredient_new = " ".join(ingredient_split)

    return ingredient_new
```

```python
def preprocess_or_ingredient(ingredient):
    
    ingredient_left_split = ingredient.split(" or ")[0]
    ingredient_right_split = ingredient.split(" or ")[-1]
    amount_list = re.findall(r'\d+', ingredient_right_split)
    if '/' in amount_list:
        slash_index = amount_list.index('/')
        division_value = float(amount_list[slash_index - 1]) / float(amount_list[slash_index + 1])
        
        del amount_list[slash_index + 1]
        del amount_list[slash_index]
        del amount_list[slash_index - 1]
        
        if len(amount_list) == 0:
            amount = division_value
        else:
            amount = division_value + float(amount_list[0])
    elif len(amount_list) == 1: 
        try:
            amount = float(amount_list[0])
        except:
            amount = amount_translator_dict[amount_list[0]]
    else:
        amount = np.nan

    unit_list = list(set(ingredient_right_split.split(" ")).intersection(units_of_food_recipe))
    if len(unit_list) > 1:
        unit = unit_list[0]
    else:
        unit = np.nan
        
    for a in amount_list:
        ingredient_right_split = ingredient_right_split.replace(a, "")
    for u in unit_list:
        ingredient_right_split = ingredient_right_split.replace(u, "")
    ingredient_right_split = filter(None, ingredient_right_split.split(" "))
    food_list = []
    for word in ingredient_right_split    :
        if word not in food_format and word not in food_size:
            food_list.append(word)
            
    if len(food_list) >= 2:
        food = " ".join(food_list)
    elif len(food_list) >= 1:
        food = food_list[0]
    else:
        food = np.nan
    
    return ingredient_left_split, amount, unit, food
```

```python
def find_nearest(array, value):
    array = np.asarray(array)
    idx = (np.abs(array - value)).argmin()
    return array[idx]

def preprocess_recipe_instructions_ingredient(ingredient, instructions):
    instructions = " ".join(instructions)
    instructions_split = instructions.split(".")
    
    food_unit_list = []
    
    for sentence in instructions_split:
        if ingredient in sentence:
            sentence_split = sentence.split(" ")
            food_unit_list = list(set(sentence_split).intersection(units_of_food_recipe))
            food_unit_list_index = []
            for food_unit in food_unit_list:
                food_unit_list_index.append(sentence_split.index(food_unit))
            food_index = sentence_split.index(ingredient.split(" ")[0])
            
    
    food_itr = 0
    if len(food_unit_list)== 0:
        ingredient_split = ingredient.split(" ")
        for sub in ingredient_split: 
            for sentence in instructions_split:
                if sub in sentence:
                    if food_itr == 0:
                        sentence_split = sentence.split(" ")
                        food_unit_list = list(set(sentence_split).intersection(units_of_food_recipe))
                        food_unit_list_index = []
                        for food_unit in food_unit_list:
                            food_unit_list_index.append(sentence_split.index(food_unit))
                        food_index = sentence_split.index(sub)
                        food_itr += 1
    
    food_unit_array = np.asarray(food_unit_list_index)
    food_unit_distance = food_index - np.asarray(food_unit_list_index)
    food_unit_distance_list = food_unit_distance.tolist()
    value = find_nearest(food_unit_distance, 1)
    new_unit = sentence_split[food_unit_list_index[food_unit_distance_list.index(value)]]
    
    return new_unit
```

```python
def preprocess_grab_parenthesis(ingredient):
    inside_parenthesis = re.search('\(([^)]+)', ingredient).group(1)
    
    amount_list = re.findall(r'\s*(\d+|[/+*-])', inside_parenthesis)
    if '/' in amount_list:
        slash_index = amount_list.index('/')
        division_value = float(amount_list[slash_index - 1]) / float(amount_list[slash_index + 1])
        
        del amount_list[slash_index + 1]
        del amount_list[slash_index]
        del amount_list[slash_index - 1]
        
        if len(amount_list) == 0:
            amount = division_value
        else:
            amount = division_value + float(amount_list[0])
    elif len(amount_list) == 1:        
        try:
            amount = float(amount_list[0])
        except:
            amount = amount_translator_dict[amount_list[0]]
    else:
        amount = np.nan
        
    unit_list = list(set(inside_parenthesis.split(" ")).intersection(units_of_food_recipe))
    if len(unit_list) > 0:
        unit = list(set(inside_parenthesis.split(" ")).intersection(units_of_food_recipe))[0]
    else:
        unit = np.nan

    
    for a in amount_list:
        inside_parenthesis = inside_parenthesis.replace(a, "")
    for u in unit_list:
        inside_parenthesis = inside_parenthesis.replace(u, "")
        
    
    inside_parenthesis_split = filter(None, inside_parenthesis.split(" "))
    food_list = []
    for word in inside_parenthesis_split    :
        if word not in food_format and word not in food_size:
            food_list.append(word)
    
    if len(food_list) >= 2:
        food = " ".join(food_list)
    elif len(food_list) >= 1:
        food = food_list[0]
    else:
        food = np.nan
    
    ingredient = re.sub("[\(\[].*?[\)\]]", "", ingredient)
    ingredient = ingredient.replace("  ", " ")
    
    return ingredient, amount, unit, food
```

```python
def preprocess_split_ingredient(ingredient):
    
    split_item_list = list(set(ingredient.split(" ")).intersection(transition_words))
    if len(split_item_list) > 0:        
        ingredient = ingredient.split(split_item_list[0])[0]
    if 'recipe follows' in ingredient:
        ingredient = ingredient.replace('recipe follows', '')
        
            
    amount_list = re.findall(r'\s*(\d+|[/+*-])', ingredient)
    
    
    
    if len(amount_list) > 0:
        if '/' in amount_list:
            slash_index = amount_list.index('/')
            division_value = float(amount_list[slash_index - 1]) / float(amount_list[slash_index + 1])

            del amount_list[slash_index + 1]
            del amount_list[slash_index]
            del amount_list[slash_index - 1]

            if slash_index == 1:
                amount = division_value
            else:
                amount = division_value + float(amount_list[0])
        else:
            try:
                amount = float(amount_list[0])
            except:
                amount = amount_translator_dict[amount_list[0]]
    else:
        amount = 1
        
    split_unit_list = list(set(ingredient.split(" ")).intersection(units_of_food_recipe))
    if len(split_unit_list) == 0:
        unit = 'whole'
        food_list = re.findall(r'\b[a-z]+\b', ingredient)
        food = " ".join(food_list).lstrip(" ")
    else:
        unit = split_unit_list[0]
        food = ingredient.split(split_unit_list[0])[-1].lstrip(" ")
            
    return amount, unit, food
```

```python
def preprocess_container_split_ingredient(ingredient):
    
    split_transition_list = list(set(ingredient.split(" ")).intersection(transition_words))
    if len(split_transition_list) > 0:        
        ingredient = ingredient.split(split_transition_list[0])[0]
    
    ingredient_split = ingredient.split(' can ')
    if len(ingredient_split) != 2:
        ingredient_split = ingredient.split(' package ')
    food = ingredient_split[-1]

    amount_list = re.findall(r'\s*(\d+|[./+*-])', ingredient_split[0])
        
    if len(amount_list) == 2:
        try:
            amount = float(amount_list[0]) * float(amount_list[1])
        except:
            amount = amount_translator_dict[amount_list[0]] * float(amount_list[1])
    elif len(amount_list) == 4:
        try:
            amount = float(amount_list[0]) * float(amount_list[1] + amount_list[2] + amount_list[3])
        except:
            amount = amount_translator_dict[amount_list[0]] * float(amount_list[1])        
    else:
        amount = amount_list[0]
    
    unit = re.sub(r'[^a-zA-Z]+', '', ingredient_split[0])
    
    if unit != 'oz' or unit != "oz" or unit != "oz." or unit != "ounce" or unit != "ounces":
        if not unit:
            unit = 'oz'
            amount = 12
        
    return amount, unit, food
```

```python
def preprocess_juice_and_zest_ingredient(ingredient):
    ingredient_split = ingredient.split(" ")
    amount_list = re.findall(r'\d+', ingredient)

    
    number_index = ingredient_split.index(amount_list[0])
    
    if len(amount_list) > 0:
        if '/' in amount_list:
            slash_index = amount_list.index('/')
            division_value = float(amount_list[slash_index - 1]) / float(amount_list[slash_index + 1])

            del amount_list[slash_index + 1]
            del amount_list[slash_index]
            del amount_list[slash_index - 1]

            if slash_index == 1:
                amount = division_value
            else:
                amount = division_value + float(amount_list[0])
        else:
            try:
                amount = float(amount_list[0])
            except:
                amount = amount_translator_dict[amount_list[0]]
    else:
        amount = 1
    
    food = ingredient_split[number_index+1:][0]
    
    
    return amount, food
```

```python
def preprocess_number_translator_ingredient(ingredient):
        
    ingredient_split = ingredient.split(" ")
    itr = 0
    while itr < len(ingredient_split):
        if ingredient_split[itr] in amount_translator_dict.keys():            
            ingredient_split[itr] = str(amount_translator_dict[ingredient_split[itr]])
        
        itr += 1
    ingredient = " ".join(ingredient_split)
    return ingredient

    
```
